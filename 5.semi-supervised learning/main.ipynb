```python
import pandas as pd
from run import main
from Utils.config import Config

import warnings
warnings.filterwarnings(action='ignore')
```

```python
results_df = pd.DataFrame(columns=['Seed','Strategy','n_1','n_2','n_3','acc_1','acc_2','acc_3'])

args = Config()

labels_list = [500, 2500, 5000, 12500, 25000]
```

Self-Training with threshold strategy

```python
for num_label in labels_list:
    args.num_l = num_label
    args.strategy = 'threshold'
    acc, iter_acc, label_rate = main(args)
    crt_dict = {'Seed':args.seed,
                'Strategy':args.strategy,
                'n_1':label_rate[0],
                'n_2':label_rate[1],
                'n_3':label_rate[2],
                'acc_1':iter_acc[0],
                'acc_2':iter_acc[1],
                'acc_3':iter_acc[2]}
    results_df = pd.concat([results_df, pd.DataFrame([crt_dict])])
```

Self-Training with top-k strategy

```python
for num_label in labels_list:
    args.num_l = num_label
    args.strategy = 'top_k'
    acc, iter_acc, label_rate = main(args)
    crt_dict = {'Seed':args.seed,
                'Strategy':args.strategy,
                'n_1':label_rate[0],
                'n_2':label_rate[1],
                'n_3':label_rate[2],
                'acc_1':iter_acc[0],
                'acc_2':iter_acc[1],
                'acc_3':iter_acc[2]}
    results_df = pd.concat([results_df, pd.DataFrame([crt_dict])])
```

Self-Training with both strategy

```python
for num_label in labels_list:
    args.num_l = num_label
    args.strategy = 'both'
    acc, iter_acc, label_rate = main(args)
    crt_dict = {'Seed':args.seed,
                'Strategy':args.strategy,
                'n_1':label_rate[0],
                'n_2':label_rate[1],
                'n_3':label_rate[2],
                'acc_1':iter_acc[0],
                'acc_2':iter_acc[1],
                'acc_3':iter_acc[2]}
    results_df = pd.concat([results_df, pd.DataFrame([crt_dict])])
```

```python
results_df = results_df.reset_index()
```

```python
results_df.drop(['index'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(data=results_df[results_df['Strategy']=='threshold'], x='n_1',y='acc_1', label='Sup.')
sns.lineplot(data=results_df[results_df['Strategy']=='threshold'], x='n_1',y='acc_3', label='threshold')
sns.lineplot(data=results_df[results_df['Strategy']=='top_k'], x='n_1',y='acc_3', label='top-k')
sns.lineplot(data=results_df[results_df['Strategy']=='both'], x='n_1',y='acc_3', label='both')

plt.xlabel('Number of inital labeled data')
plt.ylabel('Accuracy')
plt.title('Test Accuray of each strategy & supervised learning')
```

```python
sns.barplot(data=results_df[results_df['Strategy']=='threshold'].T.iloc[2:5,0:1]-500,
            x=['iter1','iter2','iter3'],y=0, label='threshold')

plt.xlabel('Iteration')
plt.ylabel('Number of pseudo labeled data')
plt.title('Pseudo labeled data per iteration of threshold strategy')
plt.ylim([0,6000])
```

```python
sns.barplot(data=results_df[results_df['Strategy']=='top_k'].T.iloc[2:5,0:1]-500,
            x=['iter1','iter2','iter3'],y=5, label='top-k')

plt.xlabel('Iteration')
plt.ylabel('Number of pseudo labeled data')
plt.title('Pseudo labeled data per iteration of top-k strategy')
plt.ylim([0,6000])
```
